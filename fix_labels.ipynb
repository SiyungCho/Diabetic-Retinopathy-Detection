```python
import pandas as pd
import os
```

```python
#create a list of all the image paths inside of the folder reviewed_images
image_paths = []
for root, dirs, files in os.walk('QMIND2024_Dataset/reviewed_images'):
    for file in files:
        if file.endswith(".jpg"):
            image_paths.append(os.path.join(root, file))
print(image_paths)
```

```python
#strip the string 'QMIND2024_Dataset/reviewed_images\\' from the beginning of each path
image_paths = [path.replace('QMIND2024_Dataset/reviewed_images\\', '') for path in image_paths]
print(image_paths)
```

```python
paths_to_delete = []
for path in image_paths:
    paths_to_delete.append('QMIND2024_Dataset/new_dataset_images/'+path)
    paths_to_delete.append('QMIND2024_Dataset/enhanced_images/'+path)
print(paths_to_delete)
```

```python
# Read the data
df1 = pd.read_csv('QMIND2024_Dataset/enhanced_dataset.csv')

# Remove the rows
df1 = df1[~df1['filename'].isin(paths_to_delete)]

# Write the data back
df1.to_csv('QMIND2024_Dataset/enhanced_dataset.csv', index=False)
```

```python
# Read the data
df2 = pd.read_csv('QMIND2024_Dataset/new_dataset.csv')

# Remove the rows
df2 = df2[~df2['filename'].isin(paths_to_delete)]

# Write the data back
df2.to_csv('QMIND2024_Dataset/new_dataset.csv', index=False)
```

```python
# Read the data
df3 = pd.read_csv('QMIND2024_Dataset/combined_dataset.csv')

# Remove the rows
df3 = df3[~df3['filename'].isin(paths_to_delete)]

# Write the data back
df3.to_csv('QMIND2024_Dataset/combined_dataset.csv', index=False)
```

```python
def print_unique_counts(unique_counts):
    plt.bar('No DR' , unique_counts[0], label= 'No DR')
    plt.bar('Non Proliferative' , unique_counts[1], label = 'Non Proliferative')
    plt.bar('Proliferative' , unique_counts[2], label = 'Proliferative')
    plt.xlabel("Diagnosis")
    plt.ylabel("No. of entries")
    plt.title("Unique Counts")
    plt.show()
```

```python
#sample the dataset to get 100 images from each class
sampled_df = balanced_df.groupby('label').apply(lambda x: x.sample(100)).reset_index(drop=True)
sampled_df.to_csv('C:/Users/siyun/OneDrive/Desktop/QMIND 2024/QMIND2024_Dataset/sampled_dataset.csv', index=False)
```

```python
#create a folder which creates a copy of the images in the sampled dataset from the original folders called enhanced_images and new_dataset_images and puts them in a folder called sampled_images
import shutil
import os

# Create a new directory
base_dir = 'C:/Users/siyun/OneDrive/Desktop/QMIND 2024/QMIND2024_Dataset/sampled_images/'

# Create a new directory for each class
no_dr_dir = os.path.join(base_dir, 'No_DR')
non_proliferative_dir = os.path.join(base_dir, 'Non_Proliferative')
proliferative_dir = os.path.join(base_dir, 'Proliferative')

# No DR
no_dr_list = sampled_df[sampled_df['label'] == 0]['filename'].tolist()
for image in no_dr_list:
    shutil.copy(image, no_dr_dir)

# Non Proliferative
non_proliferative_list = sampled_df[sampled_df['label'] == 1]['filename'].tolist()
for image in non_proliferative_list:
    shutil.copy(image, non_proliferative_dir)

# Proliferative
proliferative_list = sampled_df[sampled_df['label'] == 2]['filename'].tolist()
for image in proliferative_list:
    shutil.copy(image, proliferative_dir)
```

```python
# Create a new directory for combined images
combined_dir = os.path.join(base_dir, 'combined_sampled_images')
os.makedirs(combined_dir, exist_ok=True)

# Copy all images from the three directories into the combined directory
for dir in [no_dr_dir, non_proliferative_dir, proliferative_dir]:
    for file_name in os.listdir(dir):
        shutil.copy(os.path.join(dir, file_name), combined_dir)
```
